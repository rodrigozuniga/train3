```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```
x = np.linspace(0, 50, 1000)
plt.figure(figsize=(25,10))
plt.plot(x, np.sin(x),color="red")
plt.plot(x, np.cos(x),color="blue")
plt.plot(x, np.cos(2*x),color="green")
plt.plot(x, np.cos(0.5*x),color="black")
plt.plot(x, 2*np.cos(x),color="black")

plt.grid(color="orange")
plt.title("Trigonometric Functions \n cos(x) and sin(x)")
plt.show()

plt.figure(figsize=(25,10))
plt.plot(x, np.tan(x),color="red")
plt.xlim(0,50)
plt.ylim(-5,5)
plt.xlabel("Value of X")
plt.ylabel("Value of tan(X)")
plt.title("Trig.Functions:\n Tangent of X")
plt.grid(color="cyan")
plt.show()

```

```
x = np.linspace(0, 10, 1000)
y = np.cos(x)
z = np.sin(x)
plt.figure(figsize=(250,200))

fig, (gr1,gr2,gr3) = plt.subplots(3,1,figsize=(15,15))
fig.suptitle('Vertically stacked subplots')
plt.title('cos(x)')
gr1.plot(x,y,color="red")
gr2.plot(x, z,color="blue")
plt.ylim(-10,10)
gr3.plot(x,np.tan(x),color="green")

gr1.title.set_text('cos(x)')
gr2.title.set_text('sin(x)')
gr3.title.set_text('tan(x)')
```

```
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)
fig, axs = plt.subplots(3)
fig.suptitle('Vertically stacked subplots')
axs[0].plot(x, y)
axs[1].plot(x, -y)
```
